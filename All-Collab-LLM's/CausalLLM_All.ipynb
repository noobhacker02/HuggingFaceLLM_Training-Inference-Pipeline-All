```python
model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
from datasets import load_dataset

# Example: use local text file or a dataset
dataset = load_dataset("wikitext", "wikitext-2-raw-v1")
train_dataset = dataset["train"]
eval_dataset = dataset["validation"]
```

```python
dataset['test']
```

```python
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token #gpt 2 doesnt come with padding in token so we add it manually later
```

```python
def tokenization_pro(data):
  return tokenizer(data["text"], truncation=True, padding="max_length", max_length=128)
```

```python
tokenized_train = train_dataset.map(tokenization_pro, batched=True)
tokenized_eval = eval_dataset.map(tokenization_pro, batched=True)
```

```python
tokenized_train
```

```python
tokenized_train
```

```python
tokenized_train.set_format("torch", columns=["input_ids", "attention_mask"])
tokenized_eval.set_format("torch", columns=["input_ids", "attention_mask"])
```

```python
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False  # Causal LM → no masking
)
```

```python
training_args = TrainingArguments(
    output_dir="./causal_lm_model",
    overwrite_output_dir=True,
    eval_strategy="steps", #eval means evaulating how good or bad so strat is like epochs or steps
    eval_steps=500, #that number of strat
    save_steps=500, #save model when
    logging_steps=100, #logg wehn steps
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    num_train_epochs=1,  # increase for real training
    save_total_limit=2,
    fp16=torch.cuda.is_available(),  # use GPU if available
    push_to_hub=False,
    report_to="none",
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_eval,
    data_collator=data_collator,
)
```

```python
trainer.train()
```

```python
trainer.save_model("./Causal_LLM_model-customfinetune_wiki")
tokenizer.save_pretrained("./Causal_LLM_model-customfinetune_wiki")
```
